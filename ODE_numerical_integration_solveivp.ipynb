For the equations $$\frac{dX}{dt} = \alpha - X/\tau$$
$$\frac{dY}{dt} = X - Y/\tau$$


```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

#parameters
alpha = 1
tau = 1

#initial conditions
ic = [0,0]
t0 = 0

#time to terminate simulation
tmax = 50

def derivs(t,var): # here we define the right-hand side of the equations
    ''' Input:
                            - t, a dummy parameter, requitred by solve_ivp
                            - var - an array witrh the values of variables X, Y at a givent time t
        Output:
                            -dvar derivatives defined by our diff. equations, i.e. the RHS expressions in the ODEs
    '''
    dvar = np.zeros(2)
    dvar[0] = alpha - var[0]/tau
    dvar[1] = var[0] - var[1]/tau
    return dvar

# Here we use SciPy's function sove_ivp , the solution (we called it sol below sol) is an object. 
# The time is stored in sol.t, and variables in sol.y
sol = solve_ivp(derivs,[t0, tmax], ic)
```

```python
plt.plot(sol.t, sol.y[0,:], sol.y[1,:])
plt.legend(['X(t)', 'Y(t)'])
```
