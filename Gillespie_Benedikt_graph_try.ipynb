```python
# data structure to represent graphs
class Graph:
    def __init__(self, vertices):
        self.V = vertices  # number of vertices
        self.E = []  # list to store graph edges

    def add_edge(self, u, v):
        self.E.append((u, v))  # add edge to the graph

    def remove_edge(self, u, v):
        if (u, v) in self.E:
            self.E.remove((u, v))  # remove edge from the graph
        elif (v, u) in self.E:
            self.E.remove((v, u))  # remove edge from the graph
    
    def connected_components(self):
        visited = [False] * self.V
        components = []
        def dfs(v, component):
            visited[v] = True
            component.append(v)
            for u, w in self.E:
                if u == v and not visited[w]:
                    dfs(w, component)
                elif w == v and not visited[u]:
                    dfs(u, component)
        for i in range(self.V):
            if not visited[i]:
                component = []
                dfs(i, component)
                components.append(component)
        return components

    # get the sites for possible attachment of monomers/polymers (where the degree of the vertex is less than 3)
    def get_attachment_sites(self):
        sites = []
        for v in range(self.V):
            degree = sum(1 for u, w in self.E if u == v or w == v)
            if degree < 3:
                sites.append(v)
        return sites


class Polymer:
    def __init__(self, graph):
        self.graph = graph
```
